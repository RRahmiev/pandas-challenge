```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = school_data["school_name"].count()
total_students = student_data["student_name"].count()
total_budget = school_data["budget"].sum()
math_avg = student_data["math_score"].mean()
reading_avg = student_data["reading_score"].mean()

passed_math_counts = (student_data["math_score"]>=70).sum()
passed_math = passed_math_counts/total_students *100

passed_reading_counts = (student_data["reading_score"]>=70).sum()
passed_reading = passed_reading_counts/total_students * 100

combined_pass_count = sum((student_data["reading_score"]>=70) & (student_data["math_score"]>=70))
combined_pass = combined_pass_count/total_students * 100
```

```python
district_sumamry_df = pd.DataFrame({"Total Schools" : [total_schools],
                                    "Total Students" : [total_students],
                                    "Total Budget" : [total_budget],
                                    "Average Math Score" : [math_avg],
                                    "Average Reading Score" : [reading_avg],
                                    "% Passing Math" : [passed_math],
                                    "% Passing Reading" : [passed_reading],
                                    "% Overall Passing" : [combined_pass]})

#district_summary_df["Total Budget"] = district_summary_df["Total Budget"].astype(float).map("${:,.2f}".format)
#I get an error when I attempt to format the results unfortunately, which is why I left the command as a comment.
district_sumamry_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
school_data_complete_group = school_data_complete.groupby(["school_name"])
mean_schools = school_data_complete_group.mean()
mean_df = pd.DataFrame(mean_schools, index=mean_schools.index)
```

```python
#Retrieving data from 'mean_df'
avg_math = mean_df["math_score"]
avg_reading = mean_df["reading_score"]
total_sch_students = mean_df["size"]
total_sch_budget = mean_df["budget"]

#Calculating 'Per student budget'
per_st_budget = total_sch_budget/total_sch_students

#Calculating the % passing math per school and storing it in a series
test1 = school_data_complete.loc[school_data_complete["math_score"]>=70]
test2 = test1.groupby("school_name")
test3 = test2.count()
test12 = pd.DataFrame(test3)

passing_math = test12["math_score"]/mean_df["size"]*100

#Calculating the % passing reading per school and storing it in a series
test4 = school_data_complete.loc[school_data_complete["reading_score"]>=70]
test5 = test4.groupby("school_name")
test6 = test5.count()
test11 = pd.DataFrame(test6)

passing_reading = test11["reading_score"]/mean_df["size"]*100

#Calculating the % overal pass per school and storing it in a series
test7 = school_data_complete.loc[(school_data_complete["reading_score"]>=70) & (school_data_complete["math_score"]>=70)]
test8 = test7.groupby("school_name")
test9 = test8.count()
test10 = pd.DataFrame(test9)

overal_pass = test10["reading_score"]/mean_df["size"]*100

#Retrieving the type of each school and storing it in a series
school_type_series = school_data_complete.groupby("school_name")
school_type = school_type_series["type"].unique()
```

```python
school_summary_df = pd.DataFrame({"School Type" : (school_type),
                      "Total Students" : (total_sch_students),
                      "Total School Budget" : (total_sch_budget),
                      "Per Student Budget" : (per_st_budget),
                      "Average Math Score" : (avg_math),
                      "Average Reading Score" : (avg_reading),
                      "% Passing Math" : (passing_math),
                      "% Passing Reading" : (passing_reading),
                      "% Overall Passing" : (overal_pass)})
school_summary_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

#For the binning I am using the dataframe from the 'School Summary' section of the assignment:
school_summary_df["Spending Ranges (Per Student)"] = pd.cut(school_summary_df["Per Student Budget"], bins=spending_bins, labels=labels, include_lowest=True)
school_spending_df = school_summary_df

spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]
```

```python
spending_summary = pd.DataFrame({"Average Math Score" : (spending_math_scores),
                                     "Average Reading Score" : (spending_reading_scores),
                                     "% Passing Math" : (spending_passing_math),
                                     "% Passing Reading" : (spending_passing_reading),
                                "% Overall Passing" : (overall_passing_spending)})

#Formatting results to 2 decimal points:
spending_summary["Average Math Score"] = spending_summary["Average Math Score"].astype(float).map("{:,.2f}".format)
spending_summary["Average Reading Score"] = spending_summary["Average Reading Score"].astype(float).map("{:,.2f}".format)
spending_summary["% Passing Math"] = spending_summary["% Passing Math"].astype(float).map("{:,.2f}".format)
spending_summary["% Passing Reading"] = spending_summary["% Passing Reading"].astype(float).map("{:,.2f}".format)
spending_summary["% Overall Passing"] = spending_summary["% Overall Passing"].astype(float).map("{:,.2f}".format)

spending_summary
```
